```python
df = pd.read_excel("성적 처리.xlsx", sheet_name = 'Sheet1')
df
```

```python
# 총합
# 방법1 : df['총합'] = df['국어'] + df['수학'] + df['영어'] +  df['사회'] + df['과학'] : 쌉노가다
# 방법2
df['총합'] = df.iloc[:,2:].sum(axis = 1) #이게 훨씬 스마트
df
```

```python
df['평균'] = df['총합'] / 5
df
```

```python
# min : 3위가 2명, 공동 순위는 모두 3위로 순위 부여, 다음 순위는 5위
df['순위'] = df['평균'].rank(method='min', ascending = False) # 랭킹 매기는 함수. 
df # ascending의 의미는 표에서 정렬한 것을 보이는것이 아니라 숫자를 내림차순으로 순위로 매겨준다는 뜻
```

```python
# max : 3위가 2명, 공동 순위는 모두 4위로 순위 부여, 다음 순위는 4위
df['순위'] = df['평균'].rank(method='max', ascending = False) # 랭킹 매기는 함수
df # ascending의 의미는 표에서 정렬한 것을 보이는것이 아니라 숫자를 내림차순으로 순위로 매겨준다는 뜻
```

```python
# first : 3위가 2명, 처음 발견한 애를 3위로 순위 부여, 다음 발견 순위는 4위
df['순위'] = df['평균'].rank(method='first', ascending = False) # 랭킹 매기는 함수
df # ascending의 의미는 표에서 정렬한 것을 보이는것이 아니라 숫자를 내림차순으로 순위로 매겨준다는 뜻
```

```python
# dence : 3위가 2명, 공동 순위 3위를 둘 다 3위로 순위 부여, 다음 발견 순위는 4위
df['순위'] = df['평균'].rank(method='dense', ascending = False) # 랭킹 매기는 함수
df # ascending의 의미는 표에서 정렬한 것을 보이는것이 아니라 숫자를 내림차순으로 순위로 매겨준다는 뜻
```
