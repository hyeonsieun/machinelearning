```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import pickle
```

```python
f = open("fish.dat", "rb") # rb는 이진법을 읽겠다는 뜻
fish = pickle.load(f)
f.close()
```

```python
fish
```

```python
fish.keys()
```

```python
# fish에는 도미 / 빙어의 정보가 있다.
# 도미의 길이, 도미의 무게, 빙어의 길이, 빙어의 무게가 있다.
# 생선의 길이와 무게를 보고 도미인지 빙어인지 예측해보자
```

```python
bream_length = fish['bream_length']
bream_weight = fish['bream_weight']

smelt_length = fish['smelt_length']
smelt_weight = fish['smelt_weight']
```

```python
plt.scatter(bream_length, bream_weight)
```

```python
plt.scatter(smelt_length, smelt_weight)
```

```python
plt.scatter(bream_length, bream_weight)
plt.scatter(smelt_length, smelt_weight)
plt.legend(['bream', 'smelt'])
```

```python
length = bream_length + smelt_length
weight = bream_weight + smelt_weight
```

```python
len(bream_length)
```

```python
len(bream_weight) # 도미 데이터 개수 확인
```

```python
len(smelt_length)
```

```python
len(smelt_weight) # 빙어 데이터 개수 확인
```

```python
len(length)
```

```python
data = []
for i in range(len(length)):
    data.append([length[i], weight[i]])
print(data) # 문제지를 만들어준것, 빙어와 도미의 데이터를 2차원으로 만들어주기 위한 리스트이다.
```

```python
target = [1] * 35 + [0] * 14 # 정답지를 만들어준것. 1은 도미, 0은 빙어를 의미한다.
```

```python
print(target)
```

```python
data = np.array(data)
target = np.array(target)
```

```python
data.shape
```

```python
target.shape
```

```python
# 새로운 데이터가 들어오면 가장 가까운 데이터 5개를 찾는다. 그 중 가장 유사한 거를 다수결로 따라간다.
# 그 머신러닝이 바로 from sklearn.neighbors import KNeighborsClassifier 이다.
```

```python
from sklearn.neighbors import KNeighborsClassifier

knn = KNeighborsClassifier() # 클래스를 정의함. 모델을 생성한다.
knn.fit(data, target) # 모델을 통해 학습하는 함수가 fit()이다. 클래스에 있는 것을 꺼내쓴다.
```

```python
knn.predict([[35, 400]]) # array([1])은 도미, array([0])은 빙어
# 얘는 2차원으로 학습을 했기 때문에 2차원 데이터를 넣어주어야 한다. 원래는 행렬을 넣어야 하지만 predict()가 리스트를 자동으로 행렬로 바꿔준다.
```

```python
from sklearn.metrics import accuracy_score
```

```python
knn.score(data, target)  # 모델의 성능(정확도)를 보고 테스트하는 두 번째 방법
```

```python
fifa["contract_until"] = fifa["contract_until"].str[-4:].astype(int)
```

```python
le = LabelEncoder()

fifa['continent'] = le.fit_transform(fifa['continent'])
fifa['position'] = le.fit_transform(fifa['position'])
fifa['prefer_foot'] = le.fit_transform(fifa['prefer_foot'])
fifa['continent'] = le.fit_transform(fifa['continent'])
fifa
```

```python
from sklearn.preprocessing import StandardScaler

data = fifa.iloc[:, :-1].to_numpy()
target = fifa.iloc[:, -1].to_numpy()
train_input, test_input, train_target, test_target = train_test_split(data, target)

ss = StandardScaler() # 표준화 함수들이 들어있는 라이브러리
train_scaled = ss.fit_transform(train_input) # 평균과 표준편차를 구해주는 함수. StandardScaler 클래스의 내부에 존재하는 함수이다.fit_transform는 train_input으로 평균과 표준편차를 구하고 기억하고 있음
test_scaled = ss.transform(test_input) # fit_transform으로 평균과 표준편차를 구하면 얘네는 기억을 하고 있어. 그걸로 test_input에 표준화를 한다. 여기에서 fit_transform을 하면 test_input으로 평균과 표준편차를 구해서 표준화를 해버리기 때문에 안좋아

box = []
for n in range(1, 101) :
    knn = KNeighborsRegressor(n_neighbors=n)
    knn.fit(train_scaled, train_target)
    score = knn.score(test_scaled, test_target)
    box.append(score)
    
plt.plot(range(1,101), box)
```
