```python
 # 내가 작성한 것 (강사 코드도 나랑 비슷)
 
kospi_box = [] # 모든 코스피 페이지 넣기

for n in range(1,kospi_page + 1) :
    url =requests.get("https://finance.naver.com/sise/sise_market_sum.naver?sosok=0&page={}".format(n)) 
    # n에 따라 회차를 변경해준다. 이 때 중괄호와 format함수를 이용한다.
    html = BeautifulSoup(url.text)
    table = html.find('table', class_ = 'type_2')
    table = pd.read_html(str(table))[0] # 리스트 안에 있어서 이상하게 보였던것 뿐. 꺼내면 다시 DataFrame으로 보인다.
    del table['N']
    del table['토론실']
    table = table[table['종목명'].notnull()]
    kospi_box.append(table)
    time.sleep(1)
```

```python
# 강사 코드 (내거랑 비슷)
from tqdm import tqdm # 저장까지 남은 시간을 표기할 수 있음! 너무 좋아

kospi_box = [] # 모든 코스피 페이지 넣기

for n in tqdm(range(1,kospi_page + 1)) :
    url =requests.get("https://finance.naver.com/sise/sise_market_sum.naver?sosok=0&page={}".format(n)) 
    # n에 따라 회차를 변경해준다. 이 때 중괄호와 format함수를 이용한다.
    html = BeautifulSoup(url.text)
    table = html.find('table', class_ = 'type_2')
    table = pd.read_html(str(table))[0] # 리스트 안에 있어서 이상하게 보였던것 뿐. 꺼내면 다시 DataFrame으로 보인다.
    del table['N']
    del table['토론실']
    table = table[table['종목명'].notnull()]
    table['종목'] = ['KOSPI'] * len(table) # 새로운 열 추가하는 기능! 까먹지 말자.
    kospi_box.append(table)
    time.sleep(1)
```

```python
pd.concat(kospi_box, ignore_index = True) # 열 이름을 기준으로 합친다. 위에서 만든 kospi 테이블들은 열 이름이 모두 동일하므로 그냥 합치면 됨.
                                            # ignore_index = True는 행 번호를 맞춰준다. 이걸 안하면 원래의 index를 그대로 유지해버림 
```

```python
pd.concat(kosdaq_box, ignore_index = True) # 열 이름을 기준으로 합친다. 위에서 만든 kospi 테이블들은 열 이름이 모두 동일하므로 그냥 합치면 됨.
                                            # ignore_index = True는 행 번호를 맞춰준다. 이걸 안하면 원래의 index를 그대로 유지해버림 
```

```python
# 강사님 코드 : 빈도 분석 코드
from tqdm import tqdm

title = [] # 제목은 여기에
contents = [] # 요약은 여기에

for n in tqdm(range(1, 3992, 10)) :
    url =requests.get("https://search.naver.com/search.naver?where=news&sm=tab_pge&query=%EC%97%B0%EC%84%B8%EB%8C%80%ED%95%99%EA%B5%90&sort=0&photo=0&field=0&pd=0&ds=&de=&cluster_rank=52&mynews=0&office_type=0&office_section_code=0&news_office_checked=&nso=so:r,p:all,a:all&start={}".format(n)) 
    html = BeautifulSoup(url.text)
    news = html.find("ul", class_="list_news").find_all("li", class_= "bx")

    for i in news :
        t = i.find('a', class_= 'news_tit').text
        c = i.find('div', class_='dsc_wrap').text
        title.append(t)
        contents.append(c)

    time.sleep(0.3)

```

```python
import re # 정규표현식에서 문자열에 내가 원하는 패턴을 찾는 라이브러리

a = re.findall("[가-힣]{2,}" , t) # {2,}가 붙으면 1개 이상으로 된 한글 단어를 찾는다.

dic = {}

for i in a :
    if i not in dic.keys():
        dic[i] = 1
    else:
        dic[i] += 1
dic
```
