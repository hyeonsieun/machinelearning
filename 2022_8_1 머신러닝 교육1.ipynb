```python
# 실습 : 팩토리얼함수 구현하기 (내 풀이)

def fac(n) : 
    a = 1
    while (n != 1) : 
        a *= n
        n = n-1
    return a
fac(4)
```

```python
# 실습 : 팩토리얼함수 구현하기 (강사 풀이)

def fac(n) : 
    a = 1
    for i in range(1, n+1):
        a *= i
    return a
fac(5)
```

```python
# 재귀함수 : 내가 나를 부르는 함수

def fac(n) : 
    if n == 1:
        return 1
    else :
        return n * fac(n-1)
fac(5)
```

```python
# 피보나치 수열 : 0, 1, 1, 2, 3, 5, 8, 13, 21 .........
# fibo(5) -> 3, fibo(7) -> 8
# 내 풀이 => 강사 풀이와도 완전 동일
def fibo(n) :
    if n == 1:
        return 0
    elif n == 2:
        return 1
    else:
        return fibo(n-1)+ fibo(n-2)
fibo(7)
# 이 코드는 문제가 있음. 너무 느려.....
```

```python
# 재귀함수 비효율성 해결법 : 딕셔너리 활용

count = 0
dic = { 1 : 0, 2: 1}   # dic에 초기값 추가한다. 즉, 첫번째항과 두번째항 선언함

def fibo(n) :
    global count
    count += 1
    
    if n in dic.keys():                 # 딕셔너리에 있으면 굳이 또 계산하지 말고 dic에 있는거 꺼내 쓴다.
        return dic[n]           
    else:
        dic[n] = fibo(n-1)+ fibo(n-2)
        return dic[n]                   # 딕셔너리에 없으면 dic에 추가해서 두번 계산할 필요가 없게끔 해준다. 효율성 극대화됨
```
