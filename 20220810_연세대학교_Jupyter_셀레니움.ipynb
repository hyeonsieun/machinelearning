```python
driver = webdriver.Chrome()
driver.get("https://shopping.naver.com/fresh/directfarm/home")
driver.implicitly_wait(10) # 페이지 다뜰 때 까지 최대 10초 / HTML
time.sleep(5)

total = []

last_height = driver.execute_script("return document.body.scrollHeight")

while True:
    
    source = driver.page_source
    html = BeautifulSoup(source)

    for i in html.find_all('li', class_ = "_1OKV3UfxuI"):
        title = i.find('strong', class_ = "_1lxNNH5gfD").text
        price = i.find('span', class_ = '_3MzrqSUlwR').text
        box = [title, price]
        if box not in total:
            total.append(box)
    
    print("{}개 상품 데이터 저장 완료".format(len(total)))
    
    driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
    time.sleep(5)
    
    new_height = driver.execute_script("return document.body.scrollHeight")
    
    if new_height == last_height:
        break
    else:
        last_height = new_height
```
